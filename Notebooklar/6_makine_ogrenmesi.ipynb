# **Makine Öğrenmesi**

## **Makine Öğrenmesine Giriş**

**Makine öğrenmesi,** bilgisayarların insanlara benzer şekilde öğrenmesini sağlamak amacıyla çeşitli algoritma ve tekniiklerin geliştirilmesi için çalışılan bilimsel çalışma alanıdır.

## Terminoloji

**Bağımlı Değişken:** Başka değişkenlere bağımlı olarak değerinin değiştiği değişkenlerdir.

**Bağımsız Değişken:** Başka değişkenlerden bağımsız bir şekilde değeri olan değişkenlerdir.

**Gözetimli Öğrenme:** Veri seti içerisinde ilgilenilen yanıt değişkeni var ise yani bağımsız değişkenlerin birlikteki etkileşimleri sonucunda ortaya çıkardıkları durumun ne olduğu bilgisi var ise bu durum gözetimli (denetimli/supervised) öğrenme olarak adlandırılır.

**Gözetimsiz Öğrenme:** Veri seti içerisinde ilgilenilen yanıt değişkeni yok ise yani bağımsız değişkenlerin birlikteki etkileşimleri sonucunda ortaya çıkardıkları durumun ne olduğu bilgisi yok ise bu durum gözetimsiz (denetimsiz/unsupervised) öğrenme olarak adlandırılır.

**Regresyon:** Bağımlı değişken sürekli ise regresyon problemidir.

**Sınıflandırma:** Bağımlı değişken sürekli değil ise sınıflandırma problemidir.

**Train:** Algoritmaya veri setini öğretmek amacıyla kullanılan orjinal veri setinin bir parçasıdır.

**Test:** Algoritmaya öğretilen veri setinin başarısını ölçmek için kullanılan veri setinin bir başka parçasıdır.

**Değişken Seçimi:** Bağımlı değişken üzerinde etkisi olmayan değişkenlerin elenmesi, etkili olan değişkenlerin seçilmesidir.

**Model Seçimi:** Model seçimi regresyon için açıklanabilirlik oranı ve RMSE benzeri bir değer veya sınıflandırma için doğru sınıflandırma oranı benzeri bir değer göz önünde bulundurularak oluşabilecek değişken kombinasyonları ile oluşturulan modeller arasından en iyi modelin seçilmesi veya kurulan birbirinden farklı modeller arasından model seçimidir.

**Aşırı Öğrenme:** Modelin train verisini ezberleyerek başarısını çok yüksek oranlara taşıması, yeni bir veri modele verildiğinde ise başarısının çok düşük olması olayıdır.

**Deterministik Modeller:** Değişkenler arasında kesin bir ilişki olduğunu varsayan modellere deterministik modeller denir.

**Stokastik Modeller:** Değişkenler arasındaki ilişkinin belli bir hata ile oluştuğunu varsayan, makine öğrenmesinin ilgilendiği modellerdir.

**Doğrusal Modeller:** Değişkenler arasındaki ilişkinin doğrusal olduğu modellerdir.

**Doğrusal Olmayan Modeller:** Değişkenler arasındaki ilişkinin doğrusal olmadığı modellerdir.

## Model Doğrulama Yöntemleri

Modellerin ürettiği sonuçların doğru değerlendirilmesi çalışmalarıdır.

### Holdout Yöntemi

Orijinal veri seti eğitim ve test seti olmak üzere iki parçaya bölünür. Eğitim seti ile kurulan modelin performansı test seti ile ölçülür.

### K-Katlı Çapraz Doğrulama Yöntemi

Orijinal veri seti eğitim ve test seti olmak üzere iki parçaya bölünür. Ardından eğitim seti k adet parçaya ayrılır. Belirlenen alt kümelerden biri dışarıda bırakılarak kalan veri seti ile bir model oluşturulur. Dışarıda kalan alt küme ile model test edilir. k adet iterasyonda yöntem sonladırılır.

### Leave One Out Yöntemi

K-katlı çapraz doğrulama yönteminin özel bir halidir. Bu yöntemde alt küme sayısı gözlem sayısına eşittir. Model (n-1) adet örnek ile eğitilir ve kalan 1 adet örnek ile test edilir.

### Booststrap Yöntemi

Orijinal veri setinden yerine koymalı örnekleme yöntemiyle örnekler çekilir. Bu örnekler üzerinden her birinin test hatasının ortalaması alınarak modelin test hatası belirlenir.

**R-squared değeri;** bağımsız değişkenin bağımlı değişkendeki değişkenliği açıklama başarısıdır.

**Adj. R-squared değeri;** R-squared değerinin hassaslığı düşük versiyonudur. R-squared değeri modele değişken eklendikçe artmaya meyillidir. Dolayısıyla modele eklenen ilgisiz değişkenler de bu değeri artıracaktır. Modele eklenen bu anlamsız değişkenler ile R-squared değişkeninin artması başarıyı artırmaz. Adj. R-squared değeri bu konuda daha az hassas olduğu için analizin bu değer üzerinden yapılması daha uygundur.

**F-statistic değeri;** modelin anlamlılığının anlaşılması için oluşturulan test istatistiği değeridir.

**Prob (F-statistic) değeri;** modelin anlamlılığına ileşkin değerin p-value değeridir. Bu örnek için değerler incelendiğinde model anlamlıdır.

**coef;** katsayı değerleridir. Bu örnekte b0 = 7.0326, b1 = 0.0475'dir.

**t ve P>|t|;** katsayıların anlamlılığına ilişkin p-value değerleridir. Bu modelde katsayılar anlamlıdır.

**[0.025 0.975];** katsayıların güven aralığı değerleridir.

**b0 değeri (7.0326);** diğer değişkenlerde bir değişiklik olmadığında y'nin beklenen değeri, yani ortalama satıştır.

**b1 değeri (0.0475);** TV harcamalarında 1 birimlik artış olduğunda satışlarda 0.0475'lik bir artış olur.

### **K-Means - Teori**

Amaç gözlemleri birbirlerine olan benzerliklerine göre kümelere ayırmaktır.

**K-Means Adımları;**

Adım 1: Küme sayısı belirlenir.

Adım 2: Rastgele k merkez seçilir.

Adım 3: Her gözlem için k merkezlere uzaklıklar hesaplanır ve gözlemler kendisine en yakın k merkezine atanır.

Adım 4: Her gözlem en yakın olduğu merkeze yani kümeye atanır.

Adım 5: Atama işlemlerinden sonra oluşan kümeler için tekrar merkez hesaplamaları yapılır.

Adım 6: Bu işlem belirlenen bir iterasyon adedince tekrar edilir ve küme içi hata kareler toplamlarının minimum olduğu durumdaki gözlemlerin kümelenme yapısı nihai kümelenme olarak seçilir.
