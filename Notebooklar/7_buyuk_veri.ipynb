# **Büyük Veri**

## **Büyük Veri Nedir?**

Geleneksel yöntemlerle işlenemeyen verilere büyük veri denir. Bu açıklamaya göre büyük veri bir problemdir. Eğer büyük veri bir problem ise bu problemin çözümü birden fazla bilgisayarın bir araya gelerek tek bir bilgisayar gibi hareket etmesidir. Yani bir işi yapmak için birden fazla bilgisayarın tek bir bilgisayar gibi davranmasıdır.

Ortaya çıkan veri işleme güçlüğü verinin hacmi (volume), çeşitliliği (variety) ve hızı (velocity) ile alakalıdır. Büyük veriyi ifade eden özellikler hacim, çeşitlilik ve hızdır. **Çok yaygın bir yanılgı bu özelliklerin birlikte olduğunda ancak veriye büyük veri denilebileceğidir.**

## **Apache Hadoop**

Apache Hadoop açık kaynak kodlu, güvenilir, ölçeklenebilir paralel hesaplama yazılımı projesidir. Büyük veri teknolojilerinin temelini oluşturur. Geleneksel yöntemler ile etkin olarak işlenmesi mümkün olmayan verilerin işlenebilmesine olanak sağlamaktadır. Bir bilgisayar kümesinin belirli bir işi yapmak için tek bir bilgisayar gibi birlikte hareket etmesini sağlamaktadır.

### Apache Hadoop Bileşenleri

* Hadoop Common: Tüm diğer Hadoop modellerini destekleyen ortak gereksinimlerdir.

* Hadoop Dağıtık Dosya Sistemi (HDFS): Büyük boyutlardaki veriyi dağıtık bir biçimde kontrol etmeye olanak sağyalan bir dosya sistemidir.

* Hadoop YARN: Kaynak yönetimi ve iş planlaması için kullanılır.

* Hadoop MapReduce: Aynı ağdaki dağıtık bilgisayar kümeleri üzerinde büyük veri analizi yapılabilmesi için geliştirilmiş bir programlama modelidir.

### MapReduce Nedir?

## **Apache Spark**

Apache Spark küme üzerinde hızlı ve genel amaçlı bilgi işleme sistemidir. MapReduce modelinde yer alan disk bazlı çalışma sisteminin yarattığı maliyetlerden dolayı ortaya çıkmıştır. Apache Hadoop'a göre 100 kat hızlı çalışmaktadır. Java, Scala, **Python** ve R ile uygulama geliştirilebilir.

### Apache Spark'ın Bileşenleri

* Spark Core ve RDD's: Apache Spark'ın temel yapısını oluşturan bir kütüphanedir. RDD's sistemi ile işlemlerin hepsi bitirilmeden bir RAM'de saklanır. Tüm işlemlerin bitirildiği, başka bir işlemin olmayacağı bilgisi geldiğinde, veri RAM'den diske aktarılır.

* Spark SQL: SQL veya bir dataframe api kullanarak Spark programındaki yapılandırılmış veriyi sorgulama imkânı sağlar.

* Spark MLlib: Apache Spark'ın ölçeklendirilebilir makine öğrenmesi kütüphanesidir.

* Spark Streaming: Akan verinin ölçeklenebilir, yüksek hacimli ve hata töleranslı olarak işlenmesine imkân sağlayan temel bir Spark api'ıdır.

* GraphX: Grafikler için kullanılan bir kütüphanedir.

### Dayanıklı Dağıtık Datasetler (RDD's)

Tüm işlemler yapılarak bekletilir ardından işlemlerin bittiğini ifade eden bir komut ile işlemler bitirilir. (Mutfaktan eşya getirme örneği)

### MapReduce vs RDD's

## **Ekosistemin Diğer Üyeleri**

* Flume: Yapısal olmayan veya yarı yapısal veriler Flume adı verilen bir program ile HDFS'e aktarılır.

* Sqoop (SQL-to-Hadoop): Yapısal verileri HDFS ortamına aktarmak veya HDFS'den almak için kullanılır.

* Apache Ambari ve Zookeeper: Apache Hadoop kümesinin kurulumu, yönetimi ve gözetimi için geliştirilmiş web tabanlı araçlardır. Sistem yönetim işlerinin sağlandığı bileşendir.

* Oozie: İş zamanı programlama (job scheduling) işlemleri için kullanılan bir araçtır.

* Hbase: No-SQL database olarak ifade edilmiş sütun bazlı bir veri tabanıdır.

* Pig: My-SQL tarzı işlemler yapmak için kullanılabilen bir scripting dilidir.

* Hive: Arka planda SQL komutlarını MapReduce'a çevirerek büyük veri üzerinde SQL sorgulamaları yapılabilmesini sağlayan bir programdır.

* Mahout: Hadoop'un makine öğrenmesi aracıdır.

* Kafka & Storm: Akan veride kullanılan bir programdır.

### Test-Train

```python
splits = final_df.randomSplit([0.7, 0.3])
```

```python
train_df = splits[0]
test_df = splits[1]
```

```python
train_df
```

```python
test_df
```

## **GBM ile Müşteri Terk Modellemesi**

```python
from pyspark.ml.classification import GBTClassifier
```

```python
gbm = GBTClassifier(maxIter = 10, featuresCol = "features", labelCol = "label")
```

```python
gbm_model = gbm.fit(train_df)
```

```python
y_pred = gbm_model.transform(test_df)
```

```python
y_pred
```

```python
ac = y_pred.select("label", "prediction")
```

Modelin ilkel test hatası;

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

Model tuning;

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

evaluator = BinaryClassificationEvaluator()

paramGrid = (ParamGridBuilder()
             .addGrid(gbm.maxDepth, [2, 4, 6])
             .addGrid(gbm.maxBins, [20, 30])
             .addGrid(gbm.maxIter, [10, 20])
             .build())

cv = CrossValidator(estimator= gbm, estimatorParamMaps = paramGrid, evaluator=evaluator, numFolds= 10)
```

```python
cv_model = cv.fit(train_df)
```

.transform ile tahmin yapıyoruz;

```python
y_pred = cv_model.transform(test_df)
ac = y_pred.select("label", "prediction")
```

```python
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
evaluator.evaluate(y_pred)
```

## **Bazı Diğer Modeller**

```python
from pyspark.ml.classification import LogisticRegression
loj = LogisticRegression(featuresCol = "features", labelCol = 'label', maxIter=10)
loj_model = loj.fit(train_df)
y_pred = loj_model.transform(test_df)
ac = y_pred.select("label", "prediction")
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
from pyspark.ml.classification import DecisionTreeClassifier
dt = DecisionTreeClassifier(featuresCol = 'features', labelCol = 'label', maxDepth = 3)
dt_model = dt.fit(train_df)
y_pred = dt_model.transform(test_df)
ac = y_pred.select("label", "prediction")
ac.filter(ac.label == ac.prediction).count() / ac.count()
```

```python
from pyspark.ml.classification import RandomForestClassifier
rf = RandomForestClassifier(featuresCol = 'features', labelCol = 'label')
rf_model = rf.fit(train_df)
y_pred = rf_model.transform(test_df)
ac = y_pred.select("label", "prediction")
ac.filter(ac.label == ac.prediction).count() / ac.count()
```
